```python
def count_deaf_rats(b):
    b = b.replace(' ' ,'')
    b1= b[:b.find('P')]
    b1_list = []
    while b1:
        b1_list.append(b1[:2])
        b1 = b1[2:]
    
    b2= b[b.find('P')+1:]
    b2_list = []
    while b2:
        b2_list.append(b2[:2])
        b2 = b2[2:]
    return len( [ i for i in b1_list if i != '~O']  ) + len( [ i for i in b2_list if i != 'O~']  ) 
```

```python
def SJF(jobs, index):
    time = 0
    for i in range(len(jobs)):
        if jobs[i]<jobs[index]:
            time+=jobs[i]
            jobs[i] = 0
        else:
            continue
    for j in range(index+1):
        if jobs[j]==jobs[index]:
            time+=jobs[j]
    return time
```
